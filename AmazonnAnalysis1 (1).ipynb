```python
from scipy.stats import f_oneway

first_class = table_books[table_books['Ship Mode'] == 'First Class']['Sales']
same_day = table_books[table_books['Ship Mode'] == 'Same Day']['Sales']
second_day = table_books[table_books['Ship Mode'] == 'Second Class']['Sales']
standard_class = table_books[table_books['Ship Mode'] == 'Standard Class']['Sales']

result = f_oneway(first_class, same_day, second_day, standard_class)

print(result.pvalue)
```

```python
from scipy.stats import f_oneway

first_class = table_books[table_books['Ship Mode'] == 'First Class']['Profit']
same_day = table_books[table_books['Ship Mode'] == 'Same Day']['Profit']
second_day = table_books[table_books['Ship Mode'] == 'Second Class']['Profit']
standard_class = table_books[table_books['Ship Mode'] == 'Standard Class']['Profit']

result = f_oneway(first_class, same_day, second_day, standard_class)

print(result.pvalue)
```

Does Customer Purchases Influence the Losses

```python
customer_purchase = df.groupby('Customer ID')['Discount'].mean().reset_index()
customer_purchase.head()
```

```python
customer_behavior = df.groupby(['Customer ID', 'Customer Name']).agg(
    Total_Sales=('Sales', 'sum'),
    Net_Profit=('Profit', 'sum'),
    Avg_Discount=('Discount', 'mean')
).reset_index().sort_values("Net_Profit", ascending=False)

customer_behavior.head()
```

```python
top_five = customer_behavior
top_five.head()
```

```python
top_losses = customer_behavior[customer_behavior['Net_Profit'] <= 0].sort_values("Net_Profit", ascending = True)
top_losses.head()
```

```python
customer_category = df.groupby(['Customer ID','Category','Sub-Category']).agg(
    Total_Sales=('Sales', 'sum'),
    Net_Profit=('Profit', 'sum'),
    Avg_Discount=('Discount', 'mean')
).reset_index().sort_values("Net_Profit", ascending=False)

customer_category.head()
```
